## Introduction

> Les maladies cardiovasculaires (MCV) figurent parmi les affections non transmissibles les plus meurtrières à l’échelle mondiale.Selon l'organisation mondiale de la santé, en 2019, elles ont causé près de 17,9 millions de décès, à ce jour responsables d’environ 32 % de tous les décès dans le monde.
>
> Le fardeau des MCV ne cesse de croître : de 1990 à 2021, le nombre de décès cardiovasculaires est passé d’environ 12,1 millions à 20,5 millions selon les estimations de World Heart Federation, soit une hausse d’environ 60 %.
>
> Globalement, l'institution "American heart association" estime qu’en 2021 environ 612 millions de personnes vivaient avec une maladie cardiovasculaire.
>
> En Amérique, certains pays sont particulièrement affectés. Haiti se distingue par un taux de mortalité par MCV très élevé : 427,7 décès pour 100 000 habitants, le plus élevé de la région selon les estimations de la Pan American Health Organization.
>
> Dans le contexte haïtien, les études disponibles montrent déjà des signes alarmants : une enquête urbaine à Port-au-Prince a estimé une prévalence ajustée des MCV à ~ 14,7 % dans la population étudiée (incluant l’insuffisance cardiaque, AVC, infarctus) ; l’hypertension est le facteur de risque le plus répandu, touchant 49 % des femmes et 38 % des hommes âgés de 35 à 64 ans selon l’enquête DHS de 2017.
>
> De plus, même parmi les jeunes (18-30 ans), près de 23,5 % présentent une pression artérielle élevée, ce qui suggère une progression précoce du risque cardiovasculaire selon World Heart Federation.
>
> Ces données soulignent l’importance d’une analyse sur les maladies cardiovasculaires pour comprendre les facteurs de risque, identifier les populations les plus vulnérables, et proposer des modèles prédictifs adaptés afin de guider les politiques de prévention et d’intervention.

## Compréhension des données

> Le jeu de données utilisé dans cette étude provient de la plateforme Kaggle, sous le nom de “Cardiovascular Disease Dataset”. Il regroupe des informations cliniques et comportementales collectées sur plus de 70 000 individus, utilisées pour prédire la probabilité de survenue d’une maladie cardiovasculaire.
>
> Chaque observation représente un patient et est décrite par 12 variables explicatives et une variable cible (cardio) qui indique la présence (1) ou l’absence (0) d’une maladie cardiovasculaire diagnostiquée.Chaque enregistrement correspond à un individu et comprend plusieurs indicateurs médicaux, biologiques et liés au mode de vie.

>

>De même, la variable gender est encodée numériquement :1 = femme, 2 = homme.
>
>Ainsi, bien que toutes les valeurs apparaissent sous forme d’entiers ou de nombres réels, plusieurs représentent en réalité des catégories ou des ordres logiques qu’il faudra interpréter correctement au moment de l’analyse.

### - Cleaning

```python
# renommer certaines colonnes
df.rename(columns={'ap_hi':'p_systole','ap_lo':'p_diastole','gluc':'glucose','alco':'alcool'},inplace=True)
df
```

```python
#Conversion des ages en annees
df['age']=(df['age']/365).round(0).astype(int)
df
```

```python
# correction des valeurs negatives 
pression=['p_systole','p_diastole']
df[pression]=df[pression].abs()
```

```python
# correction des valeurs abberantes au niveau des pressions arterielles
# Seuil de normalite
#Pression systolique (ap_hi) : entre 90 et 200 mmHg
#Pression diastolique (ap_lo) : entre 60 et 120 mmHg , (Pression systolique>Pression diastolique)
```

```python
#Correction systole
df.loc[(df['p_systole'] > 250) & (df['p_systole'] <= 2500), 'p_systole'] /= 10
df.loc[df['p_systole'] > 2500, 'p_systole'] /= 100
```

```python
#Correction des pressions diastoliques 
df.loc[(df['p_diastole'] > 200) & (df['p_diastole'] <= 2000), 'p_diastole'] /= 10
df.loc[df['p_diastole'] > 2000, 'p_diastole'] /= 100
```

```python
#verification des valeurs maximales en systole
print(f"la pression systolique max est: {df['p_systole'].max()} tandisque la pression diastolique max est: {df['p_diastole'].max()}")
```

```python
#verification des valeurs minimales en systole
print(f"la pression systolique minimale est: {df['p_systole'].min()} tandisque la pression diastolique minmale est: {df['p_diastole'].min()}")
```

```python
#Lors de la contraction (systole), le cœur pousse le sang dans les artères, ce qui fait monter la pression.
#Quand il se relâche (diastole), la pression redescend.
#Donc, la pression systolique est toujours plus élevée que la diastolique.

#verification de cette condition
df[df['p_systole'] < df['p_diastole']]
```

```python
# Correction : si systole < diastole et systole < 70 → systole = systole * 10
df.loc[(df['p_systole'] < df['p_diastole']) & (df['p_systole'] < 70), 'p_systole'] = df['p_systole'] * 10
df[df['p_systole'] < df['p_diastole']]
```

```python
# Inverser les valeurs quand la diastole est plus grande que la systole
mask = df['p_diastole'] > df['p_systole']
df.loc[mask, ['p_systole', 'p_diastole']] = df.loc[mask, ['p_diastole', 'p_systole']].values
```

> ### L'age et les risques cardio

```python
# Groupement par 'age' et 'cardio'
age_cardio_grouped = df.groupby(['age', 'cardio']).size().unstack(fill_value=0)
age_cardio_grouped['Total']=age_cardio_grouped.sum(axis=1)
#calcul des pourcentage
age_cardio_grouped['AVC']=(age_cardio_grouped[1]/age_cardio_grouped['Total'])*100
age_cardio_grouped['sans_AVC']=(age_cardio_grouped[0]/age_cardio_grouped['Total'])*100
```

```python
# Graphique en ligne
plt.figure(figsize=(12, 8))  
ax=age_cardio_grouped[['AVC','sans_AVC']].plot(kind='bar',stacked=True, ax=plt.gca(), color=['red', 'green'])
plt.title('Tendances des cas de risque cardiaque par âge', fontsize=14)
plt.xlabel('Âge', fontsize=14)
plt.ylabel('Poucentage', fontsize=14)
plt.legend(title='Risque cardiaque', labels=['Pas de risque', 'Risque'])
plt.grid()
plt.show();
```

> #### Observations:
>
>Le risque cardio-vasculaire augmente avec l’âge. A partir de 40-45 ans, le risque croit beaucoup plus rapide mais atteint le pic autour de 55-60 ans, c'est une periode a haut risque qui merite une attention medicale accrue et une prevention ciblee. 

> ### Genre et cardio

```python
# Calculer le pourcentage de risque par sexe
percent = df.groupby(['gender', 'cardio']).size().groupby(level=0).apply(lambda x: 100 * x / x.sum()).unstack(fill_value=0)

# Préparer les données pour le graphique
# barres côte-à-côte par sexe : 1=femme, 2=homme
percent = percent.reindex([1, 2])  # s'assure que l'ordre est correct

# Graphique à barres côte-à-côte
ax = percent.plot(kind='bar', figsize=(10,6), color=['lightblue', 'salmon'])
plt.title('Risque cardiaque selon le sexe')
plt.xlabel('Sexe (1=femme, 2=homme)')
plt.ylabel('Pourcentage')
plt.xticks(rotation=0)
plt.legend(title='Risque cardiaque', labels=['Pas de risque', 'Risque'])

#Ajout des pourcentages sur les barres
for i, gender in enumerate(percent.index):
    for j, cardio in enumerate(percent.columns):
        ax.text(i + j*0.25 - 0.12, percent.loc[gender, cardio] + 1,  # position x et y
                f"{percent.loc[gender, cardio]:.1f}%", ha='center', va='bottom', fontsize=10)

plt.show();
```

> #### Observations:
> Pour les deux groupes, la difference est relativement faible, ce qui indique qu'il n'y a pas vraiment de relation entre les risques et le sexe.

```python
# Groupement par 'cholesterol' et 'cardio'
cholesterol_cardio_grouped = df.groupby(['cholesterol', 'cardio']).size().unstack(fill_value=0)

# Calculer le pourcentage par niveau de cholestérol
cholesterol_percent = cholesterol_cardio_grouped.div(cholesterol_cardio_grouped.sum(axis=1), axis=0) * 100

# Graphique à barres côte-à-côte
ax = cholesterol_percent.plot(kind='bar', figsize=(10,6), color=['lightblue', 'salmon'])

plt.title('Risque cardiaque selon le niveau de cholestérol')
plt.xlabel('Niveau de cholestérol (1, 2, 3)')
plt.ylabel('Pourcentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Risque cardiaque', labels=['Pas de risque', 'Risque'])

# Ajouter les pourcentages sur les barres
for i, chol in enumerate(cholesterol_percent.index):
    for j, cardio in enumerate(cholesterol_percent.columns):
        ax.text(i + j*0.25 - 0.12, cholesterol_percent.loc[chol, cardio] + 1,  # position x et y
                f"{cholesterol_percent.loc[chol, cardio]:.1f}%", ha='center', va='bottom', fontsize=10)

plt.show();
```

> #### Observations:
> Il y a une relation positive entre le niveau de cholesterol (faible, moyen ou eleve) et le risque

```python
# Groupement par 'active' et 'cardio'
active_cardio_grouped = df.groupby(['active', 'cardio']).size().unstack(fill_value=0)

# Calculer le pourcentage par niveau d'activité
active_percent = active_cardio_grouped.div(active_cardio_grouped.sum(axis=1), axis=0) * 100

# Graphique à barres côte-à-côte
ax = active_percent.plot(kind='bar', figsize=(10,6), color=['lightblue', 'salmon'])

plt.title('Risque cardiaque selon le niveau d\'activité')
plt.xlabel('Niveau d\'activité (0 = inactif, 1 = actif)')
plt.ylabel('Pourcentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Risque cardiaque', labels=['Pas de risque', 'Risque'])

# Ajouter les pourcentages sur les barres
for i, act in enumerate(active_percent.index):
    for j, cardio in enumerate(active_percent.columns):
        ax.text(i + j*0.25 - 0.12, active_percent.loc[act, cardio] + 1,
                f"{active_percent.loc[act, cardio]:.1f}%", ha='center', va='bottom', fontsize=10)

plt.show();
```

```python

# Création des tranches avec p_systole et p_diastole
df['p_systole_bin'] = pd.cut(df['p_systole'], bins=[80,120,140,160,200], labels=['80-120','121-140','141-160','161-200'])
df['p_diastole_bin'] = pd.cut(df['p_diastole'], bins=[60,80,90,100,120], labels=['60-80','81-90','91-100','101-120'])

# Calculer pourcentage de risque
hi_stats = df.groupby('p_systole_bin')['cardio'].mean() * 100
lo_stats = df.groupby('p_diastole_bin')['cardio'].mean() * 100

# Création de la figure avec 2 graphiques côte-à-côte
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Graphique pression systolique
bars_hi = axes[0].bar(hi_stats.index, hi_stats, color='salmon')
axes[0].set_title('Pourcentage de risque selon la pression systolique')
axes[0].set_ylabel('Pourcentage de personnes à risque (%)')
axes[0].set_xlabel('Tranche de pression systolique')
axes[0].set_ylim(0, max(hi_stats.max(), lo_stats.max()) + 10)

# Ajouter les pourcentages sur les barres
for bar in bars_hi:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2, height + 1, f"{height:.1f}%", ha='center', va='bottom', fontsize=10)

# Graphique pression diastolique
bars_lo = axes[1].bar(lo_stats.index, lo_stats, color='lightblue')
axes[1].set_title('Pourcentage de risque selon la pression diastolique')
axes[1].set_ylabel('Pourcentage de personnes à risque (%)')
axes[1].set_xlabel('Tranche de pression diastolique')
axes[1].set_ylim(0, max(hi_stats.max(), lo_stats.max()) + 10)

# Ajouter les pourcentages sur les barres
for bar in bars_lo:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2, height + 1, f"{height:.1f}%", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show();
```
